# Merging Joining Concatination

```python

import numpy as np 
import pandas as pd 
```

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries)
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
# Meaning of "inner" join:
# Inner join returns only the rows where the key (employee_id here) is present in both DataFrames.
# If an employee_id exists in one DataFrame but not in the other → that row is excluded from the result.
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
employees
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

 # Concatination of 2 dataframes

```python

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat((df1,df2)) # column wise
```

```python
pd.concat((df1,df2),axis=1)  # row wise
```

## joining 2 data frames 

```python

df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how="right")
```

```python
df1.join(df2,how="outer")
```
